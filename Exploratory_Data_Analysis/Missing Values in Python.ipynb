## How to Treat Missing Values in Data in Python

By Default,missing values are represented with NaN(Not a number)
!!!Warning : If your dataset has 0s,99s or 999s,be sure to either drop or approximate them as you would with missing values.

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

### Figuring out what data is missing

```python
missing = np.nan
series_obj = Series(['row1','row2',missing,'row4','row6',missing,'row8'])
series_obj
```

```python
series_obj.isnull()
```

### Filling in for missing values

```python
np.random.seed(25)
df_obj = DataFrame(np.random.rand(36).reshape(6,6))
df_obj
```

```python
df_obj.loc[3:5,0] = missing
df_obj.loc[1:4,5] = missing
df_obj
```

#### Kayıp değerlerin yerine 0 değeri verme

```python
filling_df = df_obj.fillna(0)
filling_df
```

#### Kayıp değerlerin yerine istenilen sayıları verme.

```python
filled_df = df_obj.fillna({0:0.1,5:1.25})
filled_df
```

#### Kayıp değerlerin yerini en son bilinen değerlerle doldurma.

```python
fill_df = df_obj.fillna(method = 'ffill')
fill_df
```

### Counting Missing Values

```python
np.random.seed(25)
df_obj = DataFrame(np.random.rand(36).reshape(6,6))
df_obj.loc[3:5,0] = missing
df_obj.loc[1:4,5] = missing
df_obj
```

#### Her bir sütunda ne kadar kayıp değer olduğunu gösterir.

```python
df_obj.isnull().sum()
```

### Filtering out missing values

#### Kayıp değerlerin olduğu sütunu tümüyle siler.

```python
df_no_nan = df_obj.dropna(axis=1)
df_no_nan
```

```python
df_obj.dropna(how='all')
```
